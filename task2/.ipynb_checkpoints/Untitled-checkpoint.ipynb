```python
def isYellow(name):
    img=cv2.imread(name)
    img_hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # lower mask (0-10)
    lower_red = np.array([20,50,50])
    upper_red = np.array([40,255,255])
    mask0 = cv2.inRange(img_hsv, lower_red, upper_red)


    # join my masks
    mask = mask0

    # set my output img to zero everywhere except my mask
    output_img = img.copy()
    output_img[np.where(mask==0)] = 0

    # or your HSV image, which I *believe* is what you want
    output_hsv = img_hsv.copy()
    output_hsv[mask!=0] = (0,0,0)
    return np.count_nonzero(mask) > 50
```

```python
def isRed(name):
    img=cv2.imread(name)
    img_hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # lower mask (0-10)
    lower_red = np.array([0,50,50])
    upper_red = np.array([10,255,255])
    mask0 = cv2.inRange(img_hsv, lower_red, upper_red)

    # upper mask (170-180)
    lower_red = np.array([170,50,50])
    upper_red = np.array([180,255,255])
    mask1 = cv2.inRange(img_hsv, lower_red, upper_red)

    # join my masks
    mask = mask0+mask1

    # set my output img to zero everywhere except my mask
    output_img = img.copy()
    output_img[np.where(mask==0)] = 0

    # or your HSV image, which I *believe* is what you want
    output_hsv = img_hsv.copy()
    output_hsv[mask!=0] = (0,0,0)
    return np.count_nonzero(mask) > 50
    #cv2.imshow("hsvtoRGB", cv2.cvtColor(output_hsv, cv2.COLOR_HSV2BGR))
    #cv2.imshow("hsv",output_hsv )
    #cv2.imshow("orig-hsv",img_hsv)
    #cv2.imshow("img",output_img )
    #cv2.imshow("orig",img)
    #cv2.waitKey(0)
```

```python
def isBlack(name):
    img=cv2.imread(name)
    img_hsv=cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # lower mask (0-10)
    lower_red = np.array([0,0,0])
    upper_red = np.array([255,255,0])
    mask0 = cv2.inRange(img_hsv, lower_red, upper_red)

    # join my masks
    mask = mask0

    # set my output img to zero everywhere except my mask
    output_img = img.copy()
    output_img[np.where(mask==0)] = 0

    # or your HSV image, which I *believe* is what you want
    output_hsv = img_hsv.copy()
    output_hsv[mask!=0] = (0,0,0)
    return np.count_nonzero(mask) > np.size(mask)/2
```

```python
def convertBlack(name):
    img=cv2.imread(name)
    cv2.imwrite(name,cv2.bitwise_not(img))
```

```python
for i in range(11,900):
    name = "frame"+str(i)+".jpg"
    current = "./frames/"+name
    if(os.path.isfile(current) == False):
        continue
    if (isRed(current)):
        os.rename(current, "./R/"+name)
        continue
    if (isBlue(current)):
        os.rename(current, "./U/"+name)
        continue
    if(isYellow(current)):
        os.rename(current,"./Y/"+name)
        continue
    if(isGreen(current)):
        os.rename(current,"./G/"+name)
        continue
    if(isBlack(current)):
        os.rename(current,"./BlackBackground/"+name)
        convertBlack("./BlackBackground/"+name)
    else:
        os.rename(current,"./WhiteBackground/"+name)

        
```

```python
def recognize(i):
    char = 'P'
    chars = ['A','B','I','W','O','V']
    try:
        char = pytesseract.image_to_string(Image.open("./BlackBackground/frame"+str(i)+'.jpg'),config ='--psm 10 --oem 1 --user-words blackWords.txt')
    except FileNotFoundError:
        print(i)
    if char in chars:
        try:
            os.rename("./BlackBackground/frame"+str(i)+'.jpg',char+"/frame"+str(i)+".jpg")
        except:
            o=2
            print(i,char)
    else:
        try:
            os.rename("./BlackBackground/frame"+str(i)+'.jpg',"coudnt/frame"+str(i)+".jpg")
        except:
            o=2
            print("KAK",i,char)
```
